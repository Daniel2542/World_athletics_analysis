# Top 3 competitors of all time per event_type

```python

top_competitors_per_event_type = field_event.groupby("event_type")["competitor"].value_counts().groupby(level=0).head(3)
top_competitors_per_event_type
```

Top 3 competitors of all time per event type per gender

```python
top_competitors_per_event_type_gender = field_event.groupby(["event_type", "gender"])["competitor"].value_counts().groupby(level=[0, 1]).head(3)
top_competitors_per_event_type_gender
```

# Top 3 competitors of all time per category

```python
top_competitors_per_category = field_event.groupby("category")[["competitor", "all_time_rank"]].value_counts().groupby(level=0).head(3)
top_competitors_per_category
```

# Top 3 competitors of all time per category per gender

```python
top_competitors_per_category_gender = field_event.groupby(["category", "gender"])["competitor"].value_counts().groupby(level=[0, 1]).head(3)
top_competitors_per_category_gender
```

# Youngest winners per event type

```python
field_event["age"].isna().sum()
```

```python
field_event["age"] = field_event["age"].fillna(field_event["age"].mean())
```

```python
youngest_winners = field_event.loc[field_event.groupby("event_type")["age"].idxmin()]
youngest_winners
```

# Oldest participants per event type

```python
oldest_participants = field_event.loc[field_event.groupby("event_type")["age"].idxmax()]
oldest_participants
```

# oldest winners

```python
winners = field_event[field_event["pos"] == "1"]
oldest_winners = winners.loc[winners.groupby("event_type")["age"].idxmax()]
oldest_winners
```

# Athlete with the most performances/participations per event type

```python
top_participants = field_event.groupby("event_type")["competitor"].value_counts().groupby(level=0).idxmax()
top_participants
```

# Total number of unique countries with participants

```python
unique_nationality_count = field_event["nat"].nunique()
print(f"Number of nationalities participating in field events is, {unique_nationality_count}")
```

# Best country in terms of the summation of results score

```python
best_country = field_event.groupby("nat")["results_score"].sum().idxmax()
print(f"The best country in field events is, {best_country}")
```

# worst countries


```python
worst_countries = field_event.groupby("nat")["results_score"].sum().idxmin()
worst_countries
```

```python
country_scores = field_event.groupby("nat")["results_score"].sum()

sorted_worst_countries = country_scores.sort_values(ascending=True)

worst_countries = sorted_worst_countries.head(5)

worst_countries
```
